–£–ø—Ä–∞–∂–Ω–µ–Ω–∏–µ 1 –ö–∞–ª—å–∫—É–ª—è—Ç–æ—Ä

–ù–∞–ø–∏—à–∏—Ç–µ –∫–∞–ª—å–∫—É–ª—è—Ç–æ—Ä, –≤ –∫–æ—Ç–æ—Ä—ã–π –≤—ã –±—É–¥–µ—Ç–µ –≤—ã–≤–æ–¥–∏—Ç—å –≤—ã—Ä–∞–∂–µ–Ω–∏–µ —Å —Ü–µ–ª—ã–º–∏ —á–∏—Å–ª–∞–º–∏ (+, -, *, //, %) –∏ –æ–Ω –≤ –æ–∫–Ω–µ –±—É–¥–µ—Ç –≤—ã–¥–∞–≤–∞—Ç—å –æ—Ç–≤–µ—Ç –ø–æ –Ω–∞–∂–∞—Ç–∏—é –∫–Ω–æ–ø–∫–∏. –ü–æ–¥—Å–∫–∞–∑–∫–∞: –≤ —ç—Ç–æ–º –≤–∞–º –º–æ–∂–µ—Ç –ø–æ–º–æ—á—å —Ñ—É–Ω–∫—Ü–∏—è eval(). –ü—Ä–∏–º–µ—Ä –µ–µ —Ä–∞–±–æ—Ç—ã –µ—Å—Ç—å –≤ –º–∞—Ç–µ—Ä–∏–∞–ª–∞—Ö —Ç—Ä–µ—Ç—å–µ–≥–æ —Å–µ–º–∏–Ω–∞—Ä–∞.

```python
from tkinter import *
from tkinter import ttk


def calculate(*args):
    try:
        number = str(expression.get())
        meters.set(int(eval(number)))
    except ValueError:
        pass


root = Tk()
root.title("–ö–∞–ª—å–∫—É–ª—è—Ç–æ—Ä")

mainframe = ttk.Frame(root, padding="3 3 6 6")
mainframe.grid(column=0, row=0, sticky=(N, W, E, S))
root.columnconfigure(0, weight=1)
root.rowconfigure(0, weight=1)

expression = StringVar()
feet_entry = ttk.Entry(mainframe, width=7, textvariable=expression)
feet_entry.grid(column=2, row=1, sticky=(W, E))

meters = StringVar()
ttk.Label(mainframe, textvariable=meters).grid(column=2, row=2, sticky=(W, E))
ttk.Button(mainframe, text="–ü–æ—Å—á–∏—Ç–∞—Ç—å", command=calculate).grid(column=2, row=3, sticky=W)
ttk.Label(mainframe, text="–í—ã—Ä–∞–∂–µ–Ω–∏–µ:").grid(column=1, row=1, sticky=W)
ttk.Label(mainframe, text="–†–µ–∑—É–ª—å—Ç–∞—Ç:").grid(column=1, row=2, sticky=(W, E))

for child in mainframe.winfo_children():
    child.grid_configure(padx=5, pady=5)

feet_entry.focus()
root.bind("<Return>", calculate)
root.mainloop()
```

```python
from tkinter import *
from tkinter import ttk
from tkinter import colorchooser
 
root = Tk()
root.title("–¶–≤–µ—Ç–∏–∫ —Å–µ–º–∏—Ü–≤–µ—Ç–∏–∫")
root.geometry("250x200")
 
label = ttk.Label(text="Hello World")
label.pack(anchor=NW, padx=10, pady=10)
 
def select_color():
    result = colorchooser.askcolor(initialcolor="black")
    print(result)
    re_result = '#'
    for i in result[0]:
        if len(hex(255 - i)[2::]) == 2:
            re_result += str(hex(255 - i)[2::])  
        else:
            re_result += '0' + str(hex(255 - i)[2::])
    label["foreground"] = result[1]
    label['background'] = re_result
    root['background'] = re_result
 
open_button = ttk.Button(text="–í—ã–±—Ä–∞—Ç—å —Ü–≤–µ—Ç", command=select_color)
open_button.pack(anchor=NW, padx=10, pady=10)
 
root.mainloop()
```

–ú–æ–¥–∏—Ñ–∏—Ü–∏—Ä—É–π—Ç–µ –ø—Ä–∏–ª–æ–∂–µ–Ω–∏–µ —à–∞—Ä–∏–∫–∞, –ø—Ä–∏–≤–µ–¥–µ–Ω–Ω–æ–µ –≤ –ø—Ä–∏–º–µ—Ä–µ –≤—ã—à–µ, —Ç–∞–∫, —á—Ç–æ–±—ã –ø—Ä–∏ –Ω–∞–∂–∞—Ç–∏–∏ –Ω–∞ –∫–∞–Ω–≤–∞—Å —Å–æ–∑–¥–∞–≤–∞–ª—Å—è —à–∞—Ä–∏–∫ —Å–ª—É—á–∞–π–Ω–æ–≥–æ —Ü–≤–µ—Ç–∞.

```python
from tkinter import *
from tkinter import ttk
from random import randint

WIDTH = 300
HEIGHT = 200


class Ball:
    def __init__(self, color=0):
        self.R = randint(10, 50)
        self.x = randint(self.R, WIDTH - self.R)
        self.y = randint(self.R, HEIGHT - self.R)
        self.dx, self.dy = (10, 10)

        if color == 0:
            self.color = 'green'
        else:
            color = '#'
            for i in range(3):
                new = randint(0, 255)
                if len(hex(new)[2::]) == 2:
                    color += str(hex(new)[2::])  
                else:
                    color += '0' + str(hex(new)[2::])
            self.color = color

        self.ball_id = canvas.create_oval(self.x - self.R,
                                     self.y - self.R,
                                     self.x + self.R,
                                     self.y + self.R, fill=self.color)

    def move(self):
        self.x += self.dx
        self.y += self.dy
        if self.x + self.R > WIDTH or self.x - self.R <= 0:
            self.dx = -self.dx
        if self.y + self.R > HEIGHT or self.y - self.R <= 0:
            self.dy = -self.dy

    def show(self):
        canvas.move(self.ball_id, self.dx, self.dy)


def click(event, *args):
    balls.append(Ball(1))

def tick():
    for ball in balls:
        ball.move()
        ball.show()
    root.after(50, tick)


root = Tk()
root.geometry(f'{WIDTH}x{HEIGHT}')
canvas = Canvas(root)
canvas.pack()
canvas.bind('<Button-1>', click)
balls = [Ball() for i in range(5)]
tick()
root.mainloop()
```
