### Упражение 1
Написать функцию, считающую N-ое число Фибоначчи (N вводится с клавиатуры).

```python
def fibonacci(n):
    if n == 1:
        return 0
    if n in (2, 3):
        return 1
    return fibonacci(n - 1) + fibonacci(n - 2)
 

print(fibonacci(int(input())))
```

```python
def prime_multipliers(n):
   i = 2
   prime_numbers = []
   while i ** 2 <= n:
       while n % i == 0:
           prime_numbers.append(i)
           n = n / i
       i = i + 1
   if n > 1:
       prime_numbers.append(n)
   return prime_numbers


prime_multipliers(72)
```

Реализуйте функцию, генерирующую набор данных размером N, которые методом МНК аппроксимируются прямой с коэффициентами a и b. Для этого вам пригодится библиотека random и функция random.gauss().

```python
import random
def datasets(a, b):
    
    data_size = random.randint(100, 300)
    x = np.arange(0, data_size, 1.)
    delta = np.empty((data_size, 1))
    y = np.empty((data_size, 1))
    sigma = random.uniform(1, 10)
    
    mu = 0.0 # допустим

    for i in range(data_size):
        delta[i] = random.gauss(mu, sigma)
        y[i] = a + b*x[i] + delta[i]
        
    return x, y


datasets(2, 6)
```
