```python
class Vector():
    def __init__(self, x: int, y: int, z: int, mass=1):
        assert (isinstance(x, int) and isinstance(y, int) and isinstance(z, int)) or (isinstance(x, float) and isinstance(y, float) and isinstance(z, float))
        self._data = [x, y, z]
        self._x = self._data[0]
        self._y = self._data[1]
        self._z = self._data[2]
        self.len = ((self._data[0] ** 2) + (self._data[1] ** 2) +  (self._data[2] ** 2)) ** (1 / 2)

    def pere(self):
        self._x = self._data[0]
        self._y = self._data[1]
        self._z = self._data[2]
        self.len = ((self._data[0] ** 2) + (self._data[1] ** 2) +  (self._data[2] ** 2)) ** (1 / 2)

    def __add__(self, others):
        if isinstance(others, Vector):
            data_new = [self._data[i] + others._data[i] for i in range(3)]
            return Vector(data_new[0], data_new[1], data_new[2])
        elif isinstance(others, list):
            data_new = [self._data[i] + others[i] for i in range(3)]
            return Vector(data_new[0], data_new[1], data_new[2])
    
    def __sub__(self, others):
        data_new = [self._data[i] - others._data[i] for i in range(3)]
        return Vector(data_new[0], data_new[1], data_new[2])

    def __mul__(self, others):
        if isinstance(others, float) or isinstance(others, int):
            return self.non_vec_product(self._data, others)
        elif isinstance(others, Vector):
            return self.vec_product(self._data, others._data)

    def vec_product(self, me, others):
        data_new = [me[i] * others[i] for i in range(3)]
        return Vector(data_new[0], data_new[1], data_new[2])

    def non_vec_product(self, me, others):
        data_new = [me[i] * others for i in range(3)]
        return Vector(data_new[0], data_new[1], data_new[2])
    
    def center_mass(self, me, data):
        vec_cen =  me
        for i in range(len(data)):
            vec_cen = self.center_mass_2(vec_cen, data[i])
        return vec_cen
    
    def center_mass_2(self, me, other):
        me_new = me - other
        me_new = me_new * (1 / 2)
        me_new = me_new + other
        return me_new

    def get_data(self):
        return self._data
    
    def get_area(self):
        return self
```

```python
def area(v1, v2, v3):
    assert isinstance(v1, Vector) and isinstance(v2, Vector) and isinstance(v3, Vector)
    a = v1 - v2
    a = a.len
    b = v1 - v3
    b = b.len
    c = v2 - v3
    c = c.len
    p = (a + b + c) / 2
    area = (p * (p - a) * (p - b) * (p - c)) ** (1 / 2)
    return area
    
```

```python
def max_area(*args):
    if isinstance(args[0], list):
        args = (i for i in args[0])
        args = tuple(args)

        

    combinations = tuple(itertools.combinations(args, 3))
    max_of_area = 0
    for i in combinations:
        now = area(i[0], i[1], i[2])
        if max_of_area < now:
            max_of_area = now

    return max_of_area
```

```python
from random import randrange as rnd, choice
from tkinter import *
import math
import time
```

```python
class Ball():
    """ Класс Ball описывает мяч. """
    global canv, t1

    def __init__(self,x=40,y=450):
        """ Конструктор класса Ball
        Args:
        x - начальное положение мяча по горизонтали
        y - начальное положение мяча по вертикали
        """
        self.x = x
        self.y = y
        self.r = 10
        self.vx = 0
        self.vy = 0
        self.color = choice(['blue','green','red','brown'])
        self.id = canv.create_oval(self.x - self.r, self.y - self.r, self.x + self.r, self.y + self.r, fill=self.color)
        self.live = 30
        self.g = -100
        self.time = time.time()
        self.delta_time = time.time()

    def set_coords(self):
        canv.coords(self.id, self.x-self.r, self.y-self.r, self.x+self.r, self.y+self.r)

    def move(self):
        self.delta_time = time.time() - self.time
        """ Метод move описывает перемещение мяча за один кадр перерисовки. То есть, обновляет значения 
        self.x и self.y с учетом скоростей self.vx и self.vy, силы гравитации, действующей на мяч,
            и стен по краям окна (размер окна 800х600).
        """
        if self.x > (800 - self.r) and self.vx > 0:
            self.vx = -self.vx
        if self.x < (0 + self.r) and self.vx < 0:
            self.vx = -self.vx

        if self.y > (600 - self.r) and self.vy > 0:
            self.vx *= 0.5
            self.vy *= -0.5
            if abs(self.vy) < 1 :
                self.vy = 0
                self.g = 0
                self.vx = 0
                self.__del__()

        if self.y < (0 + self.r) and self.vy < 0:           
            self.vy = -self.vy
        dx = self.x
        self.x += self.vx * self.delta_time
        dx = self.x - dx

        dy = self.y
        self.vy = self.vy - self.g * self.delta_time
        self.y += self.vy * self.delta_time
        dy = self.y - dy

        try:   
            canv.move(self.id, dx, dy)
        except TclError:
            return None
        self.time = time.time()

    def hittest(self,ob):
        """ Функция проверяет сталкивалкивается ли данный обьект с целью, описываемой в обьекте ob.
        Args:
            ob: Обьект, с которым проверяется столкновение.
        Returns:
            Возвращает True в случае столкновения мяча и цели. В противном случае возвращает False.
        """
        if (((self.x - ob.x) ** 2) + ((self.y - ob.y) ** 2)) ** (1 / 2) < (self.r + ob.r):
            return True
        else:
            return False
    
    def __del__(self):
        canv.delete(self.id)
        print('уничтожен')
    
```

```python
class gun():
    """ Класс gun описывает пушку. """
    def __init__(self,x=40,y=450):
        self.f2_power = 10
        self.f2_on = 0
        self.an = 1
        self.id = canv.create_line(20,450,50,420,width=7) # FIXME: don't know how to set it...
         
    def fire2_start(self,event):
        self.f2_on = 1
 
    def fire2_end(self,event):
        """ Выстрел мячом происходит при отпускании кнопки мыши.
        Начальные значения компонент скорости мяча vx и vy зависят от положения мыши.
        """
        global balls, bullet
        bullet += 1
        new_ball = Ball()
        new_ball.r += 5
        self.an = math.atan((event.y-new_ball.y)/(event.x-new_ball.x))
        new_ball.vx = self.f2_power*math.cos(self.an) * 4
        new_ball.vy = self.f2_power*math.sin(self.an) * 4
        balls += [new_ball]
        self.f2_on = 0
        self.f2_power = 10
 
 
    def targetting (self,event=0):
        """ Прицеливание. Зависит от положения мыши.
        """
        if event:
            self.an = math.atan((event.y-450)/(event.x-20))    
        if self.f2_on:
            canv.itemconfig(self.id,fill = 'orange')
        else:
            try:
                canv.itemconfig(self.id,fill = 'black')
            except TclError:
                return False
        canv.coords(self.id, 20, 450, 20 + max(self.f2_power, 20) * math.cos(self.an), 450 + max(self.f2_power, 20) * math.sin(self.an))
         

    def power_up(self):
        if self.f2_on:
            if self.f2_power < 100:
                self.f2_power += 1
            canv.itemconfig(self.id,fill = 'orange')
        else:
            try:
                canv.itemconfig(self.id,fill = 'black')
                return False
            except TclError:
                assert False
            
```

```python
class target():
    """ Класс target описывает цель. """ 
    def __init__(self, points=0, x=40,y=450):
        self.points = points
        try:
            canv.delete(self.id_points)
        except:
            None
        self.id_points = canv.create_text(30,30,text = self.points,font = '28') 
        self.live = 1
        self.id = canv.create_oval(5,5,5,5)
         
    def new_target(self):
        """ Инициализация новой цели. """
        self.__init__(self.points)
        x = self.x = rnd(600,780)
        y = self.y = rnd(300,550)
        r = self.r = rnd(2,50)
        color = self.color = 'red'
        canv.coords(self.id, x-r,y-r,x+r,y+r)
        canv.itemconfig(self.id, fill = color)
         
    def hit(self):
        canv.delete(self.id_points)
        """ Попадание шарика в цель. """
        canv.coords(self.id,-10,-10,-10,-10)
        self.points += 1
        # self.id_points = canv.create_text(30,30,text = self.points,font = '28')
```

```python
def new_game(event=''):
    global gun, t1, screen1, balls, bullet, g1, root, exsists
    canv.itemconfig(screen1, text ='')
    t1.new_target()
    bullet = 0
    balls = []
    canv.bind('<Button-1>', g1.fire2_start)
    canv.bind('<ButtonRelease-1>', g1.fire2_end)
    canv.bind('<Motion>', g1.targetting)
 
    z = 0.03
    t1.live = 1
    while t1.live or balls:
        if not(t1.live):
            break
        assert exsists
        for b in balls:
            assert exsists
            b.move()
            if b.hittest(t1) and t1.live:
                t1.live = 0
                t1.hit()
                canv.bind('<Button-1>', '')
                canv.bind('<ButtonRelease-1>', '')
                
        canv.update()
        time.sleep(0.03)
        g1.targetting()
        g1.power_up()
    canv.itemconfig(screen1, text = 'Вы уничтожили цель за ' + str(bullet) + ' выстрелов')
    canv.delete(gun)
    root.after(750,new_game)
```

```python
exsists = 1
root = Tk()
fr = Frame(root)
root.geometry('800x600')
canv = Canvas(root, bg = 'white')
canv.pack(fill=BOTH,expand=1)

t1 = target()
screen1 = canv.create_text(400,300, text = '',font = '28')
g1 = gun()
bullet = 0
balls = []

new_game()
print(44)
def shutdown_ttk_repeat():
    print(1)
    exsists = 0
    root.eval('::ttk::CancelRepeat')
    root.destroy()

root.protocol("WM_DELETE_WINDOW", shutdown_ttk_repeat)

root.mainloop()
```
